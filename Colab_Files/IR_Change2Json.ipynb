```python
import numpy as np
import pandas as pd
import os
import shutil
import json


def directories_json(input_path, output_path):
    allSubjects = []
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    for subject in os.listdir(input_path):
        allSubjects.append(subject)

    document_counter = 0
    for index, subject in enumerate(allSubjects):
        if not os.path.exists(os.path.join(output_path, subject)):
            os.makedirs(os.path.join(output_path, subject))
        document_counter = 0

        for document in os.listdir(os.path.join(input_path, subject)):
            text = ""
            try:
                with open(
                    os.path.join(input_path, subject, document), encoding="utf-8"
                ) as f:
                    text = f.read()
            except:
                pass

            # todo ! Add processed text to our cleaned dataset

            json_file = {"text": text, "label": index}

            try:
                with open(
                    os.path.join(
                        os.path.join(output_path, subject), "%d.json" % document_counter
                    ),
                    "w",
                    encoding="utf-8",
                ) as f:
                    f.write(json.dumps(json_file, ensure_ascii=False, indent=4))
                    document_counter += 1
            except:
                pass
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split
import uuid
```

```python
processed_data ='/content/parts'
if not (
            os.path.exists(os.path.join(processed_data, "train"))
            and os.path.exists(os.path.join(processed_data, "test"))
            and os.path.exists(os.path.join(processed_data, "valid"))
        ):
            if not os.path.exists(os.path.join(processed_data, "train")):
                os.makedirs(os.path.join(processed_data, "train"))
            if not os.path.exists(os.path.join(processed_data, "valid")):
                os.makedirs(os.path.join(processed_data, "valid"))
            if not os.path.exists(os.path.join(processed_data, "test")):
                os.makedirs(os.path.join(processed_data, "test"))
            allData = []
            input_path = '/content/NewData'
            for subject in os.listdir(input_path):
                for document in os.listdir(os.path.join(input_path, subject)):
                    allData.append(os.path.join(input_path, subject, document))
                X_train_eval, X_test = train_test_split(
                    allData, test_size=0.1, random_state=20
                )
                X_train, X_valid = train_test_split(
                    X_train_eval, test_size=0.15, random_state=20
                )
                for item in allData:
                    if item in X_train:
                        shutil.move(
                            item,
                            os.path.join(processed_data, "train"),
                        )
                    elif item in X_valid:
                        shutil.move(
                            item,
                            os.path.join(processed_data, "valid"),
                        )
                    elif item in X_test:
                        shutil.move(
                            item,
                            os.path.join(processed_data, "test"),
                        )
                allData.clear()
                for train_document in os.listdir(os.path.join(processed_data, "train")):
                    src = os.path.join(os.path.join(processed_data, "train"), train_document)
                    dst = os.path.join(os.path.join(processed_data, "train"), str(uuid.uuid1()) + ".json")
                    os.rename(src, dst)
                for validation_document in os.listdir(os.path.join(processed_data, "valid")):
                    src = os.path.join(os.path.join(processed_data, "valid"), validation_document)
                    dst = os.path.join(os.path.join(processed_data, "valid"),str(uuid.uuid1()) + ".json")
                    os.rename(src, dst)
                for test_document in os.listdir(os.path.join(processed_data, "test")):
                    src = os.path.join(os.path.join(processed_data, "test"), test_document)
                    dst = os.path.join(os.path.join(processed_data, "test"), str(uuid.uuid1()) + ".json")
                    os.rename(src, dst)
```
